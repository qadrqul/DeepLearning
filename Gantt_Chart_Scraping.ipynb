```python
import zipfile
import pandas as pd
```

```python
zip_file_path = 'C:/Users/user/Downloads/archive (7).zip'
extract_path = 'covid19_data'
```

```python
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
data_path = 'covid19_data/COVID 19 Containment measures data.csv'  
df = pd.read_csv(data_path)
```

```python
df_filtered = df[['Country', 'Date Start', 'Date end intended']]
```

```python
def filter_null_end_date(df):
    return df.dropna(subset=['Date end intended'])
```

```python
df_filtered = filter_null_end_date(df_filtered)
```

```python
csv_file_path = 'filtered_gantt_chart_data.csv'
df_filtered.to_csv(csv_file_path, index=False)
```
